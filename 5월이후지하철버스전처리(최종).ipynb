### 1. 차량데이터

```python
car3 = pd.read_excel(".\\일반교통량데이터\\01월 서울시 교통량 조사자료.xlsx",sheet_name='수집지점 주소 및 좌표',encoding='cp949')
car = pd.read_excel(".\\일반교통량데이터\\0"+str(num)+"월 서울시 교통량 조사자료.xlsx",sheet_name='2020년 0'+str(num)+'월',encoding='cp949')
# 파일은 car에 해당하는 데이터 이름에서 월만 바꾸면됨. 물론 for문 설정을 하면 되지만 느려지기 때문에 패스
```

```python
car3=car3[:-2]
```

```python
type(car3['주소'][43].count('구'))
```

```python
city=[] # 자치구 만들기
for i in range(len(car3.index)):
    a=car3['주소'][i][:car3['주소'][1].find('구')+1]
    b=a[car3['주소'][1].find('시')+1:]
    if b.count('구')==0:
        b=b+'구'
    city.append(b.strip())
car3['city']=city

Car3=car3[['지점번호','city']]
```

```python
car3
```

```python
###car

#car.isnull().sum() #겁나많은데; 일단 평균으로 채우자 나중에 모델 검증할때 바꿔보자.
car.loc[1]['방향']
```

```python
car_out_mean=[]
car_in_mean=[]
for i in range(len(car.index)):
    if car.loc[i]['방향']=='유입':
        car_in_mean.append(car.loc[i]['0시':'23시'].mean())
        car_out_mean.append(0)
    elif  car.loc[i]['방향']=='유출':
        car_in_mean.append(0)
        car_out_mean.append(car.loc[i]['0시':'23시'].mean())
car['일평균유입량']=car_in_mean
car['일평균유출량']=car_out_mean
```

```python
#누락값 치환 문제. 평균값으로 채워넣자

mean_in=car['일평균유입량'].mean(axis=0)
car['일평균유입량'].fillna(mean_in, inplace=True)
mean_out=car['일평균유출량'].mean(axis=0)
car['일평균유출량'].fillna(mean_out, inplace=True)
```

```python
Car3=car3[['지점번호','city']]
Car=car[['일자','요일','방향','지점번호','일평균유입량','일평균유출량']]
Mer_Car=pd.merge(Car,Car3,on='지점번호')
Mer_Car.head()
```

```python
len(Car)
```

```python
Mer_Car['일평균량']=Mer_Car['일평균유입량']+Mer_Car['일평균유출량']
Mer_Car.head()
```

```python
Mer_Car['city'].unique()
Mer_Car=Mer_Car.drop(Mer_Car[Mer_Car['city'].isin(['구리시','하남시구','과천시구','광명시구','부천시구','김포시구','고양시 구'])].index)
```

```python
#확인
Mer_Car[Mer_Car['city'].isin(['구리시','하남시구','과천시구','광명시구','부천시구','김포시구','고양시 구'])]
```

```python
Mer_Car['city'].unique()

#Mer_Car[Mer_Car['city']=='노원구']
```

```python
carresult = Mer_Car.groupby(['일자','city'])[['일평균유입량','일평균유출량']].mean().round()
carresult
### 일반교통량 완료
```

```python
carresult.to_csv(".\\전처리완료 데이터\\교통"+str(num)+"월.csv",encoding='utf-8-sig')
```
